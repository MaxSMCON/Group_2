```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Currency Converter.csv')
df.head(5)
```

```python
df = df.melt(id_vars=["FREQ", "Frequency","REF_AREA","Reference area","CURRENCY","Currency","COLLECTION","Collection","Time Period"], var_name="Date1", value_name="Exchange Rate")
```

```python
df ['Year Recorded'] = df['Date1'].str[:4]
df.head(5)
```

```python
df_new = df.loc[(df['Year Recorded'] =='2021')]

df_new
```

```python
df_new = df_new.dropna()
```

```python
df_new
```

```python
table = pd.pivot_table(df_new,index=['Reference area'],aggfunc={'Exchange Rate':np.mean})
table
```

```python
table.to_csv('Currence_2021_Exchange.csv')
```
